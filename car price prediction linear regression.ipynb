### Getting dummy variables

```python
def modifiy_doornumber(x):
    return x.map({'two':0,'four':1})

car[['doornumber']] = car[['doornumber']].apply(modifiy_doornumber)
```

```python
def modifiy_aspiration(x):
    return x.map({'std':0,'turbo':1})

car[['aspiration']] = car[['aspiration']].apply(modifiy_aspiration)
```

```python
def modify_englocation(x):
    return x.map({'front':1,'rear':0})

car[['enginelocation']] = car[['enginelocation']].apply(modify_englocation)
```

```python
def modify_fueltyp(x):
    return x.map({'gas':0,'diesel':1})

car[['fueltype']] = car[['fueltype']].apply(modify_fueltyp)
```
